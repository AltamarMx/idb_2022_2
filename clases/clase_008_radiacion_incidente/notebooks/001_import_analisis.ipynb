```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from dateutil import parser
```

```python
f = "../osm/cubo_001/run/001_export_variable_to_csv/report_SurfaceOutsideFaceIncidentSolarRadiationRateperArea_ZoneTimestep.csv"

nombres = np.genfromtxt(f,max_rows=1,delimiter=",",dtype="U")
for i in enumerate(nombres):
    print(i)
    
nombres[0] = "t"
nombres[1] = "Is_e"
nombres[2] = "Is_n"
nombres[3] = "Is_o"
nombres[4] = "Is_s"
nombres[5] = "Is_t"
nombres
```

```python
data = pd.read_csv(f,index_col=0,parse_dates=True,names=nombres,skiprows=1)
data
```

```python
data.plot(subplots=True)
```

```python
fig, ax  = plt.subplots(figsize=(12,4))

fecha1 = parser.parse("2006-01-25")
fecha2 = fecha1 + pd.Timedelta("1D")

ax.plot(data.Is_n,label="Norte")
ax.plot(data.Is_s,label="Sur")
ax.plot(data.Is_e,label="Este")
ax.plot(data.Is_o,label="Oeste")
ax.plot(data.Is_t,label="Techo")

ax.legend()
ax.set_xlim(fecha1,fecha2)
```

```python
fig, ax  = plt.subplots(figsize=(12,4))

fecha1 = parser.parse("2006-03-01")
fecha2 = fecha1 + pd.Timedelta("1D")

ax.plot(data.Is_n,label="Norte")
ax.plot(data.Is_s,label="Sur")
ax.plot(data.Is_e,label="Este")
ax.plot(data.Is_o,label="Oeste")
ax.plot(data.Is_t,label="Techo")

ax.legend()
ax.set_xlim(fecha1,fecha2)
```

```python
fig, ax  = plt.subplots(figsize=(12,4))

fecha1 = parser.parse("2006-05-25")
fecha2 = fecha1 + pd.Timedelta("1D")
for nombre in nombres[1:]:
    ax.plot(data[nombre],label=nombre)
ax.legend()
```
