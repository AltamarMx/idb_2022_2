# Vamos a revisar el clima

```python
data.To.groupby(by=[data.index.month,data.index.hour,data.index.minute]).agg(['mean','max','min']).plot(figsize=(20,4))
```

```python
mes = data.loc['2006-04']
var = 'To'
mes[var].groupby(by=[mes.index.month,mes.index.hour,mes.index.minute]).agg(['mean','max','min']).plot(figsize=(20,4))
```

```python
mes = data.loc['2006-04']
vars = ['Id', 'Ib', 'To', 'rh', 'ws'] 
for var in vars:
    mes[var].groupby(by=[mes.index.month,mes.index.hour,mes.index.minute]).agg(['mean','max','min']).plot(figsize=(20,4))

    
```

```python
data.To.resample('D').min().resample('M').mean().plot(label='min')
data.To.resample('D').max().resample('M').mean().plot(label='max')
data.To.resample('D').mean().resample('M').mean().plot(label='mean')

plt.legend()
```
