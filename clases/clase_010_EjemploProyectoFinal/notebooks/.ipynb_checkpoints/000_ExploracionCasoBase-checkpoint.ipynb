```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def importa_casos(fTo,fTi):
    To = pd.read_csv(fTo,index_col=0,parse_dates=True)
    Ti = pd.read_csv(fTi,index_col=0,parse_dates=True)
    cb = pd.concat([To,Ti],axis=1)
    renombres = {"Environment:Site Outdoor Air Drybulb Temperature[C]":"To",
                "THERMAL ZONE 1:Zone Mean Air Temperature[C]":"T1",
                "THERMAL ZONE 2:Zone Mean Air Temperature[C]":"T2",
                "THERMAL ZONE 3:Zone Mean Air Temperature[C]":"T3",
                "THERMAL ZONE 4:Zone Mean Air Temperature[C]":"T4"}

    cb.rename(columns=renombres,inplace=True)
#     cb = cb["T1","T2","T3",""].add_prefix
    return cb
```

```python
fTo = "../osm/CB_001/run/006_export_variable_to_csv/report_SiteOutdoorAirDrybulbTemperature_ZoneTimestep.csv"
fTi = "../osm/CB_001/run/007_export_variable_to_csv/report_ZoneMeanAirTemperature_ZoneTimestep.csv"
cb = importa_casos(fTo,fTi)


fTo = "../osm/CVentana_001/run/002_export_variable_to_csv/report_SiteOutdoorAirDrybulbTemperature_ZoneTimestep.csv"
fTi = "../osm/CVentana_001/run/003_export_variable_to_csv/report_ZoneMeanAirTemperature_ZoneTimestep.csv"
cv = importa_casos(fTo,fTi)


f =  "../osm/CB_001/run/006_export_variable_to_csv/report_SiteOutdoorAirDrybulbTemperature_ZoneTimestep.csv"
To = pd.read_csv(f,index_col=0,parse_dates=True)

f = "../osm/CB_001/run/008_export_variable_to_csv/report_SiteDiffuseSolarRadiationRateperArea_ZoneTimestep.csv"
Id = pd.read_csv(f,index_col=0,parse_dates=True)

f = "../osm/CB_001/run/009_export_variable_to_csv/report_SiteDirectSolarRadiationRateperArea_ZoneTimestep.csv"
Ib = pd.read_csv(f,index_col=0,parse_dates=True)

f = "../osm/CB_001/run/010_export_variable_to_csv/report_SiteWindSpeed_ZoneTimestep.csv"
ws = pd.read_csv(f,index_col=0,parse_dates=True)

f = "../osm/CB_001/run/011_export_variable_to_csv/report_SiteOutdoorAirRelativeHumidity_ZoneTimestep.csv"
hr = pd.read_csv(f,index_col=0,parse_dates=True)
```
