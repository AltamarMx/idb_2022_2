```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

def importa_casos(fTo,fTi):
    To = pd.read_csv(fTo,index_col=0,parse_dates=True)
    Ti = pd.read_csv(fTi,index_col=0,parse_dates=True)
    cb = pd.concat([To,Ti],axis=1)
    renombres = {"Environment:Site Outdoor Air Drybulb Temperature[C]":"To",
                "THERMAL ZONE 1:Zone Mean Air Temperature[C]":"T1",
                "THERMAL ZONE 2:Zone Mean Air Temperature[C]":"T2",
                "THERMAL ZONE 3:Zone Mean Air Temperature[C]":"T3",
                "THERMAL ZONE 4:Zone Mean Air Temperature[C]":"T4"}

    cb.rename(columns=renombres,inplace=True)
#     cb = cb["T1","T2","T3",""].add_prefix
    return cb
```

```python
fTo = "../osm/CB_001/run/004_export_variable_to_csv/report_SiteOutdoorAirDrybulbTemperature_ZoneTimestep.csv"
fTi = "../osm/CB_001/run/005_export_variable_to_csv/report_ZoneMeanAirTemperature_ZoneTimestep.csv"
cb = importa_casos(fTo,fTi)
# del cb['To']
cb  = cb.add_suffix("_cb")
# cb
```
