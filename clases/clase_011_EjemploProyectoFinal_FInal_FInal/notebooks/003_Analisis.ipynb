```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
f = "../data/casos.csv"
data = pd.read_csv(f,index_col=0,parse_dates=True,names= ["date","To","Ti_cb","Ti_cv"],skiprows=1)
data.plot(subplots=True)
```

```python
Timax_cb = data.Ti_cb.resample("D").max().resample("M").mean()
Timax_cv = data.Ti_cv.resample("D").max().resample("M").mean()
```
