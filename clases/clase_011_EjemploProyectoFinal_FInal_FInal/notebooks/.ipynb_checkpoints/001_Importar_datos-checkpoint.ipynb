```python
f  = "../osm/CB_002/run/eplusout.sql"
cb = read_eplusql(f,read="data",rename=True) 
cb.rename_cols([0,1,3,4],["Id","Ib","rh","ws"])
variables = ['Id', 'Ib', 'To', 'rh', 'ws', 'Ti_PAESTE','Ti_PAOESTE','Ti_PBESTE', 'Ti_PBOESTE']
CB = cb.get_data(variables)
VPAESTE  = 10 
VPBESTE  = 20 
VPAOESTE = 30 
VPBEOSTE = 40 

Vtotal = VPAESTE + VPAOESTE
CB["Tppi"] = (CB.Ti_PAESTE * VPAESTE + CB.Ti_PAOESTE * VPAOESTE)/Vtotal # INCOMPLETO

CB = CB.add_suffix("_cb")

# CB.plot(subplots=True,figsize=(14,3));
```

```python
f  = "../osm/CVentanas_004/run/eplusout.sql"
cv = read_eplusql(f,read="data",rename=True) 
variables = ['Ti_PAESTE','Ti_PAOESTE','Ti_PBESTE', 'Ti_PBOESTE']
CV = cv.get_data(variables)

VPAESTE  = 10 
VPBESTE  = 20 
VPAOESTE = 30 
VPBEOSTE = 40 
Vtotal = VPAESTE + VPAOESTE
CV["Tppi"] =( CV.Ti_PAESTE * VPAESTE + CV.Ti_PAOESTE * VPAOESTE)/Vtotal  # INCOMPLETO

CV = CV.add_suffix("_cv")

CV.plot(subplots=True,figsize=(14,3));
```
